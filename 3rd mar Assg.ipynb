# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot:
Use: Visualize the relationship between two continuous variables.
Seaborn Function: sns.scatterplot()

Line Plot:
Use: Display the trend of a numerical variable over a continuous interval or time.
Seaborn Function: sns.lineplot()

Bar Plot:
Use: Compare the values of different categories of a categorical variable.
Seaborn Function: sns.barplot()

Histogram:
Use: Illustrate the distribution of a single numerical variable.
Seaborn Function: sns.histplot()

Heatmap:
Use: Explore the correlation between two-dimensional data.
Seaborn Function: sns.heatmap()

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'flights' is your DataFrame
# Pivot the data to create a matrix for the heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Counts by Year and Month')

# Show the plot
plt.show()
```
